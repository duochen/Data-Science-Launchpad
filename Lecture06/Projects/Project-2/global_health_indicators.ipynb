### üí° Concepts:

* Compare countries by health stats
* Detect outliers in infant mortality

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Load dataset
df = pd.read_csv("health_data.csv")
print(df.head())
```

```python
# 2. Summary statistics
print("Summary:")
print(df.describe())
```

```python
# 3. Boxplot: Outliers in Infant Mortality
sns.boxplot(x=df["Infant Mortality"])
plt.title("Outliers in Infant Mortality Rate")
plt.show()
```

## üì¶ Plot 1: **Boxplot ‚Äì Infant Mortality Rate**

### üîç What It Shows:

* A **horizontal boxplot** of the `Infant Mortality` column.
* The box represents the **interquartile range (IQR)**: the middle 50% of countries.
* The **line inside the box** is the **median**.
* **Dots or points beyond the ‚Äúwhiskers‚Äù** indicate **outliers** ‚Äî countries with unusually high infant mortality.

### ‚úÖ Interpretation:

* If the whiskers are short and a few dots lie far to the right, some countries have **much higher infant mortality** than the average.
* In this dataset, **Nigeria** and **South Africa** likely appear as outliers due to very high rates.
* This plot helps detect **inequality in health outcomes**.

```python
# 4. Correlation Heatmap
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="viridis")
plt.title("Health Indicators Correlation Matrix")
plt.show()
```

## üî• Plot 2: **Correlation Heatmap**

### üîç What It Shows:

* A **colored matrix** of pairwise correlations between numeric columns:

  * `Life Expectancy`
  * `Health Spending Per Capita`
  * `Infant Mortality`

### ‚úÖ Interpretation:

* Values range from **‚Äì1 (perfect negative)** to **+1 (perfect positive)**.
* Example insights:

  * üîº **High positive correlation** (close to +1): `Health Spending` ‚Üë ‚Üí `Life Expectancy` ‚Üë
  * üîΩ **High negative correlation** (close to ‚Äì1): `Infant Mortality` ‚Üë ‚Üí `Life Expectancy` ‚Üì
* Color helps: darker = stronger relationship.

### üí° Key Insight:

* **More health spending is generally associated with higher life expectancy and lower infant mortality.**

```python
# 5. Scatter plot: Health Spending vs Life Expectancy
sns.scatterplot(x="Health Spending Per Capita", y="Life Expectancy", data=df)
plt.title("Do Richer Countries Live Longer?")
plt.show()
```

## üí¨ Plot 3: **Scatter Plot ‚Äì Health Spending vs. Life Expectancy**

### üîç What It Shows:

* Each point represents a **country**.
* X-axis: `Health Spending Per Capita`
* Y-axis: `Life Expectancy`

### ‚úÖ Interpretation:

* A **positive upward trend** means that countries that **spend more on healthcare** tend to have **higher life expectancy**.
* Example:

  * Japan and Norway likely appear toward the top-right (high spending, long lives).
  * Nigeria and India likely appear in the lower-left (low spending, shorter lives).
