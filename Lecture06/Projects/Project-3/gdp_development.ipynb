## üìå Project 3: GDP vs Development

### üìù Description:

Analyze how GDP per capita relates to literacy, urbanization, or internet access. Great for combining economics and social factors.

### üìÅ Dataset:

Use a simplified CSV (`gdp_development.csv`):

| Country | GDP per Capita | Literacy Rate | Urban Population (%) | Internet Access (%) |
| ------- | -------------- | ------------- | -------------------- | ------------------- |
| Germany | 50000          | 99            | 77                   | 92                  |

### üí° Concepts:

* Descriptive stats and skewness
* Correlation between GDP and development indicators
* Use visual and statistical EDA to find patterns

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Load dataset
df = pd.read_csv("gdp_development.csv")
print(df.head())
```

```python
# 2. Summary stats
print(df.describe())
```

```python
# 3. Skew and distribution
print("Skewness:\n", df.skew(numeric_only=True))
```

```python
# 4. Pairplot
sns.pairplot(df)
plt.suptitle("GDP and Development Indicators", y=1.02)
plt.show()
```

## üñºÔ∏è Plot 1: `sns.pairplot(df)`

### üìä What It Shows:

* A **grid of scatterplots and histograms** showing pairwise relationships between all numeric variables.

### ‚úÖ How to Interpret:

* **Diagonal**: Histograms of each individual variable.
* **Other cells**: Scatterplots showing how one variable correlates with another.

  * Example: If the points in ‚ÄúGDP vs Internet Access‚Äù form an upward trend ‚Üí strong positive relationship.

### üß† Insight:

* Countries with **higher GDP** tend to have:

  * **Higher Internet Access**
  * **Higher Literacy Rates**
  * **More Urban Populations**

```python
# 5. Correlation Heatmap
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

## üñºÔ∏è Plot 2: `sns.heatmap(df.corr(...))`

### üìä What It Shows:

* A **correlation matrix heatmap** with numbers and colors.
* Measures **linear relationships** between all pairs of variables (from -1 to 1).

### ‚úÖ How to Interpret:

* Values close to **+1**: Strong positive correlation
* Values close to **-1**: Strong negative correlation
* **Colors**: Blue/red (strong) ‚Üí white/gray (weak)

### üß† Insight:

* You‚Äôll likely see:

  * **GDP per Capita** positively correlates with **Internet Access** and **Literacy Rate**
  * Moderate correlation with **Urban Population**

```python
# 6. Insight: GDP vs Literacy
sns.scatterplot(x="GDP per Capita", y="Literacy Rate", data=df)
plt.title("GDP vs Literacy Rate")
plt.show()
```

## üñºÔ∏è Plot 3: `sns.scatterplot(x="GDP per Capita", y="Literacy Rate")`

### üìä What It Shows:

* A **scatterplot** of GDP vs Literacy Rate for each country.

### ‚úÖ How to Interpret:

* **Upward trend**: Higher GDP ‚Üí Higher literacy.
* Look for **clusters or outliers**:

  * A country with high GDP but low literacy = possible policy issue.
  * A country with low GDP but high literacy = education success story.

### üß† Insight:

* Shows that economic development is often (but not always) associated with education access.
* Encourages critical thinking: ‚ÄúWhat causes exceptions?‚Äù
