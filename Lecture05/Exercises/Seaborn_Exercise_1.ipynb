```python
# Import packages
import seaborn as sns
%matplotlib inline
```

```python
# Load a dataset from seaborn
tips_dataset = sns.load_dataset('tips')
tips_dataset.head()
```

```python
# Use distplot - distribution plot
sns.displot(tips_dataset['total_bill'],bins=30,kde=False)
```

```python
# Estimate tip with respect to total_bill
sns.relplot(x="total_bill",y="tip",data=tips_dataset)
```

```python
# Estimate tip with respect to total_bill and sex
sns.relplot(x="total_bill",y="tip",data=tips_dataset,hue='sex')
```

```python
# Plot pairwise relationships in a datasetS
sns.pairplot(tips_dataset,hue='sex',palette='coolwarm')
```

```python
# Show the counts of observations in each categorical bin using bars.
sns.countplot(x='sex',data=tips_dataset)
```

```python
# Import pandas
import pandas as pd  # import pandas to use groupby()
tips_dataset.groupby('sex').count()
```
