## 📊 1. **Line Plot (`ax1.plot(...)`)**

### 🔍 What It Shows:

* A **line connecting data points** defined by `x` and `y`.
* Demonstrates trends or changes over time or index.

### ✅ How to Interpret:

* Look at the **slope** of the line to understand increases or decreases.
* Smooth trends are ideal for visualizing performance or growth.

## 📊 3. **Scatter Plot (`ax3.scatter(x, y)`)**

### 🔍 What It Shows:

* Each point on the plot corresponds to an `(x, y)` pair.
* Used to identify **correlations** or **clusters** between two variables.

### ✅ How to Interpret:

* A **tight, upward-sloping group** of points suggests a positive correlation.
* A **random scatter** implies no clear relationship.
* You may spot **outliers** that deviate from general patterns.

```python
# Initialize the plot
fig = plt.figure(figsize=(20,10))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

# Plot the data
ax1.bar([1,2,3],[3,4,5])
ax2.barh([0.5,1,2.5],[0,1,2])

# Show the plot
plt.show()
```

## 📊 4. **Histogram (`ax4.hist(...)`)**

### 🔍 What It Shows:

* Data **frequency distribution** by dividing values into bins (ranges).
* Helps understand the **shape** of the data: normal, skewed, uniform, etc.

### ✅ How to Interpret:

* High bars = many data points in that range.
* Useful to detect **skewness** (left/right tail) or check if data is **normally distributed**.

```python
# Ensure x and y are the same size for scatter plot
import numpy as np
x = np.array([1, 2, 3, 4])
y = np.array([10, 20, 25, 30])
# Define y if not already defined
import numpy as np
y = np.array([10, 20, 25, 30])
# Initialize the plot
fig = plt.figure()
ax1 = fig.add_subplot(131)
ax2 = fig.add_subplot(132)
ax3 = fig.add_subplot(133)

# Plot the data
ax1.bar([1,2,3],[3,4,5])
ax2.barh([0.5,1,2.5],[0,1,2])
ax2.axhline(0.45)
ax1.axvline(0.65)
ax3.scatter(x,y)

# Show the plot
plt.show()
```
